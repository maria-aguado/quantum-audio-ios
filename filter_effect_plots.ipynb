**Notebook To Draw the Filter Effect PLots**

1. Plot with two frequencies

```python
import numpy as np
import matplotlib.pyplot as plt

# Frequency axis
f = np.linspace(-10, 10, 6000)

# Centers
f1 = 3
f2 = 6

# Wider lobes -> smaller f0
f0 = 0.5

def cropped_sinc_numpy(f, fc, f0):
    x = (f - fc) * f0
    y = np.sinc(x)
    mask = np.abs(x) <= 1.0  # main lobe only
    return np.where(mask, y, np.nan)

blob1 = cropped_sinc_numpy(f, f1, f0)
blob2 = cropped_sinc_numpy(f, f2, f0)

fig, ax = plt.subplots()

ax.plot(f, blob1)
ax.plot(f, blob2)

ax.fill(f, blob1, alpha=0.3)
ax.fill(f, blob2, alpha=0.3)

# Limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 1.2)

# Move x-axis to y = 0
ax.spines["bottom"].set_position(("data", 0.0))

# Remove other spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Remove ticks
ax.set_xticks([])
ax.set_yticks([])

# Draw markers crossing the x-axis
marker_h = 0.02
ax.plot([f1, f1], [-marker_h, marker_h], color="black", linewidth=1)
ax.plot([f2, f2], [-marker_h, marker_h], color="black", linewidth=1)

# Labels BELOW the x-axis (axis coordinates for y)
ax.text(f1, 0.075, "f1", ha="center", va="top", transform=ax.get_xaxis_transform(), fontsize=18)
ax.text(f2, 0.075, "f2", ha="center", va="top", transform=ax.get_xaxis_transform(), fontsize=18)

# x-title = Frequency (Hz), y-title = Amplitude
ax.set_xlabel("Frequency (Hz)", fontsize=18)
ax.set_ylabel("Amplitude", fontsize=18)

plt.show()
```
