**Notebook To Test The Encoding & Decoding of QAP**

1. Import libraries:

```python
import quantumaudio
from quantumaudio.tools import test_signal
from quantumaudio.tools import plot
```

2. Generate signal and encode it:

```python
signal = test_signal(num_channels=1,num_samples=4)  # import test signal
print("Original Signal:")
print(signal)
print("Encoding scheme:")
scheme = quantumaudio.load_scheme('qsm', qubit_depth=5) # load the QSM scheme
print("\nEncoded Circuit:") # encoded circuit = signal encoded using the scheme selected
encoded_circuit = scheme.encode(signal,verbose=2,measure=False) # encode the signal into a quantum circuit
```

4. Decoded signal and plot comparison:

```python
decoded_signal = scheme.decode(encoded_circuit)

print("\nOriginal Signal:")
print(signal)
print("\nDecoded Signal:")
print(decoded_signal)
print("\nPlot:")
plot([signal, decoded_signal])
```
